### Cleaning the Language column

```python
from nltk.stem import SnowballStemmer
```

```python
stemmer = SnowballStemmer("english")
```

```python
def clean_language(lang):
    lang = lang.lower().strip()
    
    lang = re.sub(r'sanskritenglish|sanskrit\s*english|sanskrutam\s*aanglam|sankritamangalam', 'sanskrit, english', lang)
    lang = re.sub(r'sanskrithindi|sanskrit\s*hindi', 'sanskrit, hindi', lang)
    lang = re.sub(r'hindienglish|hindi\s*english', 'hindi, english', lang)
    lang = re.sub(r'englishtamil', 'english, tamil', lang)
    lang = re.sub(r'malayalamenglish', 'malyalam, english', lang)
    
    lang = re.sub(r'[\|/]', ',', lang)
    
    langs = [stemmer.stem(l.strip()) for l in lang.split(',') if l.strip()]
    
    return ', '.join(sorted(set(langs)))  

Almost_clean_sheet['language'] = Almost_clean_sheet['language'].apply(clean_language)
```
